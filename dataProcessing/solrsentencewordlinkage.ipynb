# Loading the words from preprocessed wordList

```python
import ast

with open('/kaggle/input/wlist123/wordList.txt', 'r') as f:
    contents = f.read()
    arr = ast.literal_eval(contents)
words = arr    
```

This function utilizes the Spacy library to calculate the similarity between a new sentence and a list of existing sentences. It returns True if there is a similar sentence found (with a similarity score above the threshold), and False otherwise. If similarity is high the sentence would not be linked to the word

```python


# Load the pre-trained model
def calculateSimilarity(sentences,new_sentence):
   



    # Calculate the similarity between the new sentence and each sentence in the list
    similarities = [nlp(new_sentence).similarity(nlp(sentence)) for sentence in sentences]
    
    for i in similarities:
        if i > 0.7:
            return True
    
    return False
   



```

The Below two blocks are responsible for getting for each words the sentence present on solr, and linking ith the word with some cleaning and lanuage processing functions as described above

```python
import json
import re
import requests

def checkIfStringHasSpecialChar(_string):
    spChars = r'[!@#$%^&*()_+\-=\[\]{};\\:"|,<.>/?]+'
    if re.search(spChars, _string):
        return True
    else:
        return False

corpus = []
sentenceDict = {}
solr_url = "http://35.223.110.79:8983/solr/mycol1/query"
mongo_url = "https://us-east-1.aws.data.mongodb-api.com/app/dictionary-eokle/endpoint/addGoogleAudio"
seen_documents = set()
count = 0;


            
```

```python
for word in words[:]:
    try:

        
            count+=1
            if count == 1 or count % 1000 == 0 : print(count)
            if checkIfStringHasSpecialChar(word):
                continue


            params = {
                'q': f'text:{word}',
                'rows': 1000,
            }
            solr_response = requests.get(solr_url, params=params).json()

            if solr_response['response']['numFound'] > 0:
                document_map = {}
                for record in solr_response['response']['docs']:
                    if record['text'][0] not in document_map:
                        document_map[record['text'][0]] = [record['source'][0], record['id']]

                docs = list(document_map.keys())
                general_examples = docs


                sentence_ids = []
                senList = []
                exmpCount= 0
                for i in general_examples:
                    if exmpCount == 20 : break
                    if len(i) < 15 or len(i) > 150 :continue     
#                     if calculateSimilarity(senList,i) : continue    
                    exmpCount+=1

                    if document_map[i][1] not in seen_documents:
                        corpus.append({"id": document_map[i][1], "text": i, "source": document_map[i][0]})
                    sentence_ids.append(document_map[i][1])
                    seen_documents.add(document_map[i][1])
                    senList.append(i)



                sentenceDict[word] = sentence_ids
#                 if count % 10000 == 0 :
#                     with open(f'corp{count}', "w") as f:
#                         # Write the dictionary to the file in JSON format
#                         json.dump(corpus, f)
#                         corpus = []
    except:
        print("error at",count)


```
